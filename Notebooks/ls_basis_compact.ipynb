## Least squares fitting using compact-supported basis functions 

Dataset source: https://bit.ly/2RICuqB

Dataset description: SOLAR2000 - Solar flux measured between February 14, 1947 and May 31, 2002. This dataset was designed to be used as a tool for modelling solar radiation in space environments as well as inputs for planetary atmospheric models.

In this project, a least squares fitted model is generated from this data set using pre-defined basis functions (cubic piecewise polynomial) with compact support. This model was generated such that the output as well as its first derivative is fitted exactly, preventing unwanted oscillations generated by sharp variations in the data.

```matlab
% Set directory and load data
wdir = 'D:\Portfolio';
addpath([wdir '\data']);

data = importdata('Five_cycle_v1_23a.txt');

flux = data.data(:,2);
time = data.data(:,1);

% Set up interval and nodes
N =20;
int = length(flux)/(N-1);
node = round([1:int:length(flux) length(flux)]);


%% Hi(x) Coefficients

% Right
for j=1:length(node)-1;
    time_diff = time(node(j+1)) - time(node(j));
    A = [time_diff^2 time_diff^3; 2*time_diff 3*time_diff^2];
    cond_end = [-1; 0];
    hi_right(:,j) = inv(A)*cond_end;
end
hi_right = [hi_right zeros(2,1)];

% Left
for j=2:length(node);
    time_diff = time(node(j-1)) - time(node(j));
    A = [time_diff^2 time_diff^3; 2*time_diff 3*time_diff^2];
    cond_end = [-1; 0];
    hi_left(:,j) = inv(A)*cond_end;
end

%% Hi(x) basis function middle points 
for j=2:length(node)-1;
    for k = 1:length(time);
        if time(k) < time(node(j)) && time(k) > time(node(j-1));
            x = time(k) - time(node(j));
            hi(j,k) = 1 + hi_left(1,j)*x^2 + hi_left(2,j)*x^3;
        elseif time(k) > time(node(j)) && time(k) < time(node(j+1));
            x = time(k) - time(node(j));
            hi(j,k) = 1 + hi_right(1,j)*x^2 + hi_right(2,j)*x^3;
        end
        hi(j,node(j)) = 1;
        hi(j,node(j+1)) = 0;
        hi(j,node(j-1)) = 0;
    end
end


% left boundary
for k = 1:length(time);
    if time(k) > time(node(1)) && time(k) < time(node(2));
        x = time(k) - time(node(1));
        hi(1,k) = 1 + hi_right(1,1)*x^2 + hi_right(2,1)*x^3;
    end
end
hi(1,node(1)) = 1;

% right boundary
for k = 1:length(time);
    if time(k) > time(node(end-1)) && time(k) < time(node(end));
        x = time(k) - time(node(end));
        hi(N,k) = 1 + hi_left(1,end)*x^2 + hi_left(2,end)*x^3;
    end
end
hi(N,node(end)) = 1;

%% Gi(x) Coefficients

% Right
for j=1:length(node)-1;
    time_diff = time(node(j+1)) - time(node(j));
    A = [time_diff time_diff^2; 2*time_diff 3*time_diff^2];
    cond_end = [-1; -1];
    gi_right(:,j) = inv(A)*cond_end;
end
hi_right = [gi_right zeros(2,1)];

% Left
for j=2:length(node);
    time_diff = time(node(j-1)) - time(node(j));
    A = [time_diff time_diff^2; 2*time_diff 3*time_diff^2];
    cond_end = [-1; -1];
    gi_left(:,j) = inv(A)*cond_end;
end

%% Gi(x) basis function middle points 
for j=2:length(node)-1;
    for k = 1:length(time);
        if time(k) < time(node(j)) && time(k) > time(node(j-1));
            x = time(k) - time(node(j));
            gi(j,k) = x + gi_left(1,j)*x^2 + gi_left(2,j)*x^3;
        elseif time(k) > time(node(j)) && time(k) < time(node(j+1));
            x = time(k) - time(node(j));
            gi(j,k) = x + gi_right(1,j)*x^2 + gi_right(2,j)*x^3;
        end
        gi(j,node(j)) = 1;
        gi(j,node(j+1)) = 0;
        gi(j,node(j-1)) = 0;
    end
end


% left boundary
for k = 1:length(time);
    if time(k) > time(node(1)) && time(k) < time(node(2));
        x = time(k) - time(node(1));
        gi(1,k) = x + gi_right(1,1)*x^2 + gi_right(2,1)*x^3;
    end
end
gi(1,node(1)) = 1;

% right boundary
for k = 1:length(time);
    if time(k) > time(node(end-1)) && time(k) < time(node(end));
        x = time(k) - time(node(end));
        gi(N,k) = x + gi_left(1,end)*x^2 + gi_left(2,end)*x^3;
    end
end
gi(N,node(end)) = 1;



%% Create fit from least squares

% Create function with alternating hi(x) and gi(x) 
for i = 1:length(node);
    Pi(2*i-1,:) = hi(i,:);
    Pi(2*i,:) = gi(i,:);
end

% Create alpha matrix for least squares
for i = 1:2*length(node);
    for j = 1:2*length(node);
        alpha(i,j) = sum(Pi(i,:).*Pi(j,:));
    end
end

% Create b vector with flux data
for k = 1:2*length(node);
    b(k) = sum(flux'.*Pi(k,:)); 
end

a = inv(alpha)*b';              % Inversion to solve for weighted coeff

% Multiply with basis functions
for k = 1:length(a);
    fx(k,:) = a(k)*Pi(k,:);
end

fx = sum(fx,1);                 % Sum over wrt. time points

sigma = sqrt(mean((flux - fx').^2)); % Root mean squared error


plot(time,flux,time,fx,'o')
xlabel('Time in Julian days'); ylabel('Solar flux (SFU)');
legend('Points','Fitted model')
title(horzcat('Solar flux with least squares model, N = ', num2str(N)))

disp(horzcat('Number of nodes = ', num2str(N)))
disp(horzcat('R^2 Error = ',num2str(sigma)))
```
